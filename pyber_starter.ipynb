```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset

fares = ride_data.groupby('city')['fare'].sum()
rides = ride_data.groupby('city')['city'].count()
avg = fares / rides
merge_table = pd.merge(city_data, pd.DataFrame({'avgfare':avg, 'rides':rides, 'fares':fares}).reset_index(), on="city")


# Display the data table for preview
#city_data.head()
merge_table.head(100)
```

## Bubble Plot of Ride Sharing Data

```python
colors = {"Urban":"coral", "Suburban":"skyblue", "Rural":"yellow"}
city_labels = ["Urban", "Suburban", "Rural"]
legend_color = ["coral", "lightblue", "yellow"]
# x, Urban
x_urban = merge_table[merge_table.type=='Urban'] ['fares']
#print(x_urban)

y_urban = merge_table[merge_table.type=='Urban'] ['avgfare']
#print(x_urban)

plt.scatter(x_urban, y_urban, marker="o", facecolors="coral", edgecolors="black",
            s=x_urban, alpha=0.75)


# x, Suburban
x_Suburban = merge_table[merge_table.type=='Suburban'] ['fares']
#print(x_urban)

y_Suburban = merge_table[merge_table.type=='Suburban'] ['avgfare']
#print(x_urban)

plt.scatter(x_Suburban, y_Suburban, marker="o", facecolors="skyblue", edgecolors="black",
            s=x_Suburban, alpha=0.75)


# x, Rural
x_Rural = merge_table[merge_table.type=='Rural'] ['fares']
#print(x_urban)

y_Rural = merge_table[merge_table.type=='Rural'] ['avgfare']
#print(x_urban)

plt.scatter(x_Rural, y_Rural, marker="o", facecolors="yellow", edgecolors="black",
            s=x_Rural, alpha=0.75)

# Incorporate the other graph properties
plt.grid(axis='both')
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")

# Create a legend
plt.text(44, 37, "Note: \n" "Circle size correlates with driver count per city.")

handlelist = [plt.plot([], marker = "o", markeredgecolor = "black", ls = "", color=color)[0] for color in legend_color]
plt.legend(city_labels, markerscale=1, scatterpoints=1)

# Save Figure
plt.savefig("Pyber_Ride.png")
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
type_df = merge_table.groupby(["type"]) 
tot_rides_per_type = type_df["city"].count()
print(tot_rides_per_type[0])
print(tot_rides_per_type)

# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
# Labels for the sections of our pie chart
labels = ["Rural", "Suburban", "Urban"]

# The values of each section of the pie chart
sizes = [tot_rides_per_type[0], tot_rides_per_type[1], tot_rides_per_type[2]]

# The colors of each section of the pie chart
colors = ["yellow", "lightblue", "coral"]

# Tells matplotlib to seperate the "Python" section from the others
explode = (0, 0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)

plt.title("% Total Rides by City Type")

# Save Figure
plt.savefig("City_Type_Rides.png")

# Show Figure
plt.show()

```
